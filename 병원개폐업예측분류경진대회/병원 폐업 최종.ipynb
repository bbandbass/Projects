손익계산서.png


*  <b><u>매출액</u></b>(영업수익) - <u>**매출원가**</u> = 매출총이익

*   매출총이익 - <u>**판관비**(**급여** + @)</u> = 영업이익

*   영업이익 + <u>**영업외수익**</u> - <u>**영업외비용**(**이자비용** + @)</u> = 법인세 차감 전 이익
*   법인세 차감 전 이익 - <u>**법인세 비용**</u> = <u>**당기순이익(Net Income)**</u> (찐 순수이익)

*   자산 총계= <u>**유동자산**(**당좌자산** + **미수금(단기)** + **재고 자산** + @)</u> + <u>**비유동자산**(**유형자산** + **기타비유동자산**(**장기미수금** + @)+@)</u>+
*   <u>**부채 총계** = **유동부채**(**단기차입금** + @)</u> + <u>**비유동부채**(**장기차입금** + @)</u>
*   <u>**순자산 총계**</u> = 자본 총계 = <u>**이익잉여금**</u> + @



---

```python
#df column별 0개수 세기 
def count(df):
    dict={}
    for column in df.columns:
        count,kind=0,0
        for n,i in enumerate(df[column]):
            if i==0:
                count+=1
                if df["instkind"][n] != "general_hospital": #병원 종류가 종합병원이 아닐경우 회계정보 공시대상이 아니다./한국보건산업진흥원피셜
                    kind +=1                                #모르니까 0값으로 써져있을 가능성 up


        dict[column]=[count,kind] #count:컬럼별 0이 있는 개수, kind: 종합병원 아닌데 컬렴별 0 이 있는 개수

    dict=sorted(dict.items(),key=lambda x:x[1],reverse=True) #value값 기준 내림차순 정렬 
    return dict
```

#Model Selection

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold

def scoring(model,feature):
    rskfold = RepeatedStratifiedKFold(n_splits=5, n_repeats=5, random_state=42)  #5개로 5번 split
    #클래스의 표본비율을 전체집합과 동일하게 가져간다.(Stratified) + repeated
    acc = cross_val_score(model,feature, y_train, scoring="accuracy", cv=rskfold).mean()
    return acc

#X_train, y_train을 각각 train/validation set으로 나눠서 검증 진행     
```

```python
models = pd.DataFrame(columns=["Model","Accuracy(Cross-validation)","Note"])
```

*   estimator(string) : classifier, regressor, pipeline
